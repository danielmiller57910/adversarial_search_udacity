## Definitions

* PROBLEM_NAME -> specific air planning problem that is being solved
* SOLUTION_NAME -> name of search algorithm, if prefixed with : then a heuristic function is applied
* ACTIONS_LENGTH -> total number of actions layers required to solve the problem
* EXPANSIONS -> total number of action and literal layers added to the graph
* GOAL_TESTS -> total checks whether current literal layer satisfies goal
* STATES -> total number of possible states identified in the literal layers of the graph

```python
def set_plot_3(df1, df2, df3, x_label, y_label, title):
    fig, axes = plt.subplots(1, 3, figsize=(20, 7))
    for i, df in enumerate([df1, df2, df3]):
        col_0, col_1 = df.columns[0], df.columns[1]
        axes[i].bar(df[col_0], df[col_1])
        axes[i].title.set_text(title + f": Problem {i + 1}")
        axes[i].tick_params(axis='x', direction='out', rotation=90)
        axes[i].set_xlabel(x_label)
        axes[i].set_ylabel(y_label)
    
    plt.show()
        
set_plot_3(
    df1_subset["EXPANSIONS"], 
    df2_subset["EXPANSIONS"],
    df3_subset["EXPANSIONS"],
    "Solution Names", 
    "Number Of Expansions",
    "Expansions By Search Algorithm")
```

## Review

### Which algorithm or algorithms would be most appropriate for planning in a very restricted domain (i.e., one that has only a few actions) and needs to operate in real time?
For small domains that contain under 100 actions, greedy algorithms that utilise heuristic functions to inform search perform significantly better than informed search algorithms that do contain heuristics. For a problem space of less than 25 actions the greedy bfs search using the h_pg_max level is most appropriate and for those with 25-75 steps the level sum heuristic performs best.

### Which algorithm or algorithms would be most appropriate for planning in very large domains (e.g., planning delivery routes for all UPS drivers in the U.S. on a given day)
Depth First Serarch Algorithms have the lowest runtime though are not necessarily appropriate as they are not guaranteed to arrive at a correct solution. The greedy best_first algorithms offer a good balance of computational efficiency and accuracy. As seen by the graph they have less goal tests, states & expansions than equivalent algorithms.
